# Introduction au Machine Learning : La Régression Linéaire

```python
#Etape 1: Importer les Librairies
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
```

```python
#Etape 2: Créer un Dataset
```

```python
np.random.seed(0)
x, y = make_regression(n_samples=100, n_features=1, noise=10)
plt.scatter(x,y)
```

```python
#Etape 3: Développer et entraîner le modèle
```

```python
model = SGDRegressor(max_iter=100, eta0=0.0001)
model.fit(x,y)
plt.scatter(x,y)
plt.plot(x, model.predict(x), c='red', lw=3)
```

```python
model = SGDRegressor(max_iter=1000, eta0=0.001)
model.fit(x,y)
plt.scatter(x,y)
plt.plot(x, model.predict(x), c='red', lw=3)
```
